# Fundamental 8. 데이터 시각화 Data Visualization

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
csv_path = os.getenv("HOME") + "/data/flights.csv"
data = pd.read_csv(csv_path)
flights_df = pd.DataFrame(data)
```

```python
flights_df
```

```python
sns.barplot(data=flights_df, x='year', y='passengers')
```

```python
sns.pointplot(data=flights_df, x='year', y='passengers')
```

```python
sns.lineplot(data=flights_df, x='year', y='passengers')
```

```python
sns.lineplot(data=flights_df, x='year', y='passengers', hue='month', palette='ch:.50')
plt.legend(bbox_to_anchor=(1.03, 1), loc=2) #legend 그래프 밖에 추가하기
```

```python
sns.distplot(flights_df['passengers'])
```

```python
pivot = flights_df.pivot(index='year', columns='month', values='passengers')
```

```python
sns.heatmap(pivot)
```

```python
sns.heatmap(pivot, linewidths=.2, annot=True, fmt="d")
```

```python
sns.heatmap(pivot, cmap="YlGnBu")
```
